# Modelling Classes:

```python
class Library:
    def __init__(self, list):
        self.books_available = list
    
    def showbooks(self):             # option-1
        for i in self.books_available:
            print(i)
    
    def lend(self,book):                #option-2
        if book in self.books_available:
            self.books_available.remove(book)
        else:
            print('Sorry!!! Book not available')
        
    def receive(self,book):                 #option-3
        if book not in self.books_available:
            self.books_available.append(book)
        else:
            print('Book already returned/available')
     
    def welcome_message(self):
        print("Welcome to the Library :")

class Customer:
        
    def borrow_book(self):             # option-2
        print('Enter the name of the book to borrow:')
        borrow_book = input()
        return borrow_book
           
            
    def return_book(self):             # option-3
        print('Enter the name of the book to return or add:')
        return_book = input()
        return return_book
    
        
```

#  Creating Objects:

```python
library = Library(['book1','book2','book3','book4','book5','book6'])
customer = Customer()
```

# Calling the Functions:

```python
library.welcome_message()
call = True
while call == True:
    
    print('\nSelect one option from below:')
    
    print(' \n 1- Show books available in the library \n 2- Borrow a book from the library \n 3- Return/Add a book to the library \n 4- Exit')
    
    option = int(input())
    
    if option == 1:
        library.showbooks()
    
    elif option == 2:
        requestd_book = customer.borrow_book()
        library.lend(requestd_book)
    
    elif option == 3:
        return_book = customer.return_book()
        library.receive(return_book)
    
    elif option != 1 and option != 2 and option != 3 and option != 4:
        print('Enter a valid option')
        
    elif option == 4:
        call = False
```
